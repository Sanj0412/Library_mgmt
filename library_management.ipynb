```python
'''
Library
'''
#Library Class: This class represents the library and contains methods for showing available books,lending books, and returning books.
# Class definition for Library
class Library:
    
    # Constructor method for Library class
   def __init__(self, books):
       self.books = books
        
    # Method to display available books in the library
   def show_avail_books(self):
       print('Our Library Can Offer You The Following Books:')
       print('================================================')
       for book, borrower in self.books.items():
           if borrower == 'Free':
               print(book)
                
     # Method to lend a book to a student
   def lend_book(self, requested_book, name):
       if self.books[requested_book] == 'Free':
           print(
               f'{requested_book} has been marked'
               f' as \'Borrowed\' by: {name}')
           self.books[requested_book] = name
           return True
       else:
           print(
               f'Sorry, the {requested_book} is currently'
               f' on loan to: {self.books[requested_book]}')
           return False
        
     # Method to return a book to the library
   def return_book(self, returned_book):
       self.books[returned_book] = 'Free'
       print(f'Thanks for returning {returned_book}') #Python f-string (formatted string literal). It's used to create a string that includes the value of the variable returned_book within the string.

#Student Class: This class represents a student and contains methods for viewing borrowed books,requesting a book, and returning a book.
## Class definition for Student
class Student:
    # Constructor method for Student class
   def __init__(self, name, library):
       self.name = name
       self.books = [] # List to track books borrowed by the student
       self.library = library # Reference to the library object
        
    # Method to view books borrowed by the student
   def view_borrowed(self):
       if not self.books:
           print('You haven\'t borrowed any books')
       else:
           for book in self.books:
               print(book)

    # Method for a student to request a book
   def request_book(self):
       book = input(
           'Enter the name of the book you\'d like to borrow >> ')
       if self.library.lend_book(book, self.name):
           self.books.append(book)

    # Method for a student to return a book to the library
   def return_book(self):
       book = input(
           'Enter the name of the book you\'d like to return >> ')
       if book in self.books:
           self.library.return_book(book)
       else:
           print('You haven\'t borrowed that book, try another...')

#create_lib() Function: This function sets up the initial library with some example books, creates a student instance, and provides a menu for users to interact with the library.
#Function to create a library and initialize a student
def create_lib():
     # Initial book inventory for the library
   books = {
       'The Last Battle': 'Free',
       'The Daughter from a wishing tree': 'Free',
       '': 'Free'
   }
   library = Library(books)  # Create a Library instance
   student_example = Student('Your Name', library)  # Create a Student instance with a name and the library reference

 # Main loop for the library menu
   while True:
       print('''
           ==========LIBRARY MENU===========
           1. Display Available Books
           2. Borrow a Book
           3. Return a Book
           4. View Your Books
           5. Exit'''
             )

       choice = int(input('Enter Choice: ')) # Get user input for menu choice
       if choice == 1:
           print()
           library.show_avail_books()  # Display available books
       elif choice == 2:
           print()
           student_example.request_book() # Allow a student to borrow a book
       elif choice == 3:
           print()
           student_example.return_book() # Allow a student to return a book
       elif choice == 4:
           print()
           student_example.view_borrowed()  # View books borrowed by the student
       elif choice == 5:
           print('Goodbye')
           exit() # Exit the program if the user chooses to do so

# Entry point of the script
if __name__ == '__main__':
   create_lib()  # Start the library management system
```
